```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itables import show
```

```python
df = pd.read_csv('../data/raw/kaggle/input/playground-series-s4e4/train.csv')
show(df)
```

```python
actual_counts = df['Rings'].value_counts().sort_index()

display(actual_counts)

plt.figure(figsize=(10, 6))
actual_counts.plot(kind='bar')
plt.title('Rings Distribution')
plt.xlabel('Rings')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```
